# Big-O: Exponential : Big O(2<SUP>N</SUP>)

## Two Examples below
- Fibonacci using Recursion
- Factorial Using Recursion

## Fibonnaci Example
- The Fibonacci example below is in true sense demonstrates - BigO(2<sup>n-1</sup>)
- Why we then call it **Big O(2**<SUP>**n**</SUP>) time complexity ?
- Because **2**<SUP>**n-1**</SUP> =  **2**<SUP>**n**</SUP>/ **2**<SUP>**1**</SUP> = **(1/2) * 2**<SUP>**n**</SUP>
- Since when calculating Time complexity we ignore all constant terms, we ignore (1/2) here


```python
def fibo(n, iterlist):
    if n <= 2:
        iterlist.append(n)
        return 1
    else:
        iterlist.append(n)
        return fibo(n-1, iterlist) + fibo(n-2, iterlist)
        
```

```python
iterlist = []
runs = [5, 10, 15, 20, 25]

for n in runs: 
    print(f"fibonacci for n = {n} is: {fibo(n, iterlist)}")
    print(f"number of runs = {len(iterlist)} BigO(2^(n-1) = BigO({2**(n-2)}) complete runlist = {iterlist} ")
    print()
```

```python
def fact(n, iterrows):
    if n == 1 or n == 0:
        iterrows.append(n)
        return 1
    else:
        iterrows.append(n)
        return n * fact(n-1, iterrows)
    
```

```python
iterlist = []
runs = [5, 10, 15, 20, 25]

for n in runs: 
    print(f"Factorial for n = {n} is: {fact(n, iterlist)}")
    print(f"number of runs = {len(iterlist)} BigO(2^(n-1) = BigO({2**(n)}) complete runlist = {iterlist} ")
    print()
```
